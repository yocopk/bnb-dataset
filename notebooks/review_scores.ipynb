```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/listings_palermo.csv')
```

```python
# Rimuove i valori NaN
ratings = df['review_scores_rating'].dropna()
```

```python
# Crea l'istogramma
plt.figure(figsize=(10, 6))
plt.hist(ratings, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuzione dei Punteggi delle Recensioni')
plt.xlabel('Punteggio')
plt.ylabel('Numero di Annunci')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
