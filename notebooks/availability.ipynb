## Distribuzione disponibilità annuale

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/listings_palermo.csv')
```

```python
# 1. Distribuzione della disponibilità annuale (availability_365)
plt.figure(figsize=(10, 5))
sns.histplot(df['availability_365'], bins=50, kde=False, color='skyblue')
plt.title('Distribuzione della Disponibilità Annuale')
plt.xlabel('Giorni Disponibili all\'Anno')
plt.ylabel('Numero di Annunci')
plt.grid(axis='y')
plt.figtext(
    0.5, -0.05,
    "La maggior parte degli annunci ha una disponibilità limitata, con molti che non sono disponibili tutto l'anno.",
    ha='center', fontsize=10, style='italic'
)
plt.tight_layout()
plt.savefig('../outputs/disponibilita_365.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Notti minime e notti massime

```python
# 2. Analisi di minimum_nights e maximum_nights
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['minimum_nights'], bins=50, color='coral')
plt.title('Distribuzione Minimum Nights')
plt.xlabel('Notti Minime')
plt.ylabel('Numero di Annunci')
plt.xlim(0, 30)

plt.subplot(1, 2, 2)
sns.histplot(df['maximum_nights'], bins=50, color='mediumseagreen')
plt.title('Distribuzione Maximum Nights')
plt.xlabel('Notti Massime')
plt.ylabel('Numero di Annunci')
plt.xlim(0, 365)

plt.figtext(
    0.5, -0.02,
    "Molti annunci richiedono un minimo di 1 notte, ma esistono anche requisiti più alti. La maggior parte dei maximum nights è elevata o illimitata.",
    ha='center', fontsize=10, style='italic'
)
plt.tight_layout()
plt.savefig('../outputs/minmax_nights.png', dpi=300, bbox_inches='tight')
plt.show()
```
