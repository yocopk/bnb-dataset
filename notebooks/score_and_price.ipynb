```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Caricamento del dataset
df = pd.read_csv('../data/processed/listings_palermo.csv')
```

```python
# Pulizia: seleziona solo righe con valori validi per prezzo e punteggio
df_scatter = df[['price', 'review_scores_rating']].dropna()
```

```python
# Pulizia della colonna 'price': rimozione del simbolo $ e delle virgole, conversione a float
df_scatter['price'] = df_scatter['price'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
# Creazione dello scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_scatter, x='price', y='review_scores_rating', alpha=0.6)
plt.title('Relazione tra Prezzo e Punteggio delle Recensioni')
plt.xlabel('Prezzo (€)')
plt.ylabel('Punteggio Medio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Rimuove outlier estremi nel prezzo
df_scatter = df_scatter[df_scatter['price'] < 500]  # soglia arbitraria
```

```python
# Creazione dello scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_scatter, x='price', y='review_scores_rating', alpha=0.6)
plt.title('Relazione tra Prezzo e Punteggio delle Recensioni')
plt.xlabel('Prezzo (€)')
plt.ylabel('Punteggio Medio')
plt.grid(True)
plt.tight_layout()
plt.show()
```
