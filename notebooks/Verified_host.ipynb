```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Caricamento del dataset
df = pd.read_csv("../data/processed/listings_palermo.csv")
```

```python
# Conta i verificati
verified_counts = df['host_identity_verified'].value_counts()
label_map = {'t': 'Verificato', 'f': 'Non Verificato'}
labels = [label_map[i] for i in verified_counts.index]
```

```python
# Grafico a torta
plt.figure(figsize=(6, 7))
patches, texts, autotexts = plt.pie(
    verified_counts,
    labels=['', ''], 
    autopct='%1.1f%%',
    startangle=90,
    colors=['lightgreen', 'salmon']
)
plt.legend(patches, labels, title="Identità Host", loc="lower right")
plt.title('Distribuzione Identità Verificata degli Host')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
