```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/listings_palermo.csv')
```

```python

counts = df['instant_bookable'].value_counts(normalize=True) * 100
counts = counts.sort_index()  # ordina True/False oppure "no"/"yes"
```

## Percentuale di annunci con prenotazione immediata

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
bars = plt.bar(counts.index, counts.values, color=['mediumseagreen', 'salmon'])
plt.ylim(0, 100)
plt.title('Percentuale Annunci Prenotazione Immediata')
plt.ylabel('Percentuale (%)')
plt.xlabel('Prenotazione Immediata')

for i, val in enumerate(counts.values):
    plt.text(i, val + 2, f'{val:.1f}%', ha='center', fontsize=12)

plt.figtext(
    0.5, -0.05,
    "Mostra la percentuale di annunci che offrono prenotazione immediata, importante per capire la rapidità di conferma possibile per l'utente.",
    ha='center', fontsize=10, style='italic'
)
plt.tight_layout()
plt.savefig('../outputs/prenotabilita_instant_bar.png', dpi=300, bbox_inches='tight')
plt.show()
```
