```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Caricamento del dataset
df = pd.read_csv("../data/processed/listings_palermo.csv")
```

```python
# Converti la colonna a numerico (in caso di valori sporchi)
df['calculated_host_listings_count'] = pd.to_numeric(df['calculated_host_listings_count'], errors='coerce')
```

```python
# Grafico
plt.figure(figsize=(8, 5))
sns.histplot(df['calculated_host_listings_count'].dropna(), bins=30, kde=False, color='skyblue')
plt.title('Distribuzione del Numero di Annunci per Host')
plt.xlabel('Numero di Annunci per Host')
plt.ylabel('Frequenza')
plt.grid(True)
plt.tight_layout()
plt.show()
```
